```python
import lightgbm as lgb
import numpy as np
from sklearn.datasets import load_iris

iris = sklearn.datasets.load_iris()
X = iris.data

# mmake all features categorical
X *= 10

ds = lgb.Dataset(
    X,
    iris.target,
    feature_name=iris.feature_names,
    categorical_feature=[0, 1, 2, 3],
    params={
        "min_data_in_bin": 5,
        "min_data_per_group": 1
    }
)

bst = lgb.train(
    params={
        'num_leaves': 7,
        'objective': 'multiclass',
        'num_classes': 3,
        'min_data_in_leaf': 1,
    },
    train_set=ds,
    num_boost_round=3
)

mod = bst.dump_model()
example_case = X[[0]]
```

```python
lgb.create_tree_digraph(bst, example_case=example_case, tree_index=1)
```
